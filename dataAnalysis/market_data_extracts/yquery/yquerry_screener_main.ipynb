```python
## creating screener groups for quotes screens
target_strings = ['_au$', '_uk$', '_ca$', '_in$', '_eu$' , '_gb$', '_us$', '_br$','_fr$','_hk$','_it$','_nz$','_sg$',
                  '_asia$', '_europe$', '_americas$']

top_strings = ["^top_", "^undervalued_" , "^most_" ]

etfs_string = "_etfs$"

funds_string = "_funds$"

top_string = "^top_"

non_us_qscreen_list = []
non_us_rscreen_list = []

etf_qscreen_list = []
etf_rscreen_list = []

top_qscreen_list = []
top_rscreen_list = []

remaining_qscreen_list = []
remaining_rscreen_list = []

funds_qscreen_list = []
funds_rscreen_list = []

filter_quotes_data_screens = []

all_quotes_screens=  list(pd.read_csv(qfile)['screen_name'])

#pattern = r"(?:" + "|".join(target_strings) + r")\b"  # Regular expression with word boundaries
#pattern = re.compile(r"|".join(target_strings))
filter_non_us_pattern = re.compile("|".join(f"{target}$" for target in target_strings))
filter_etfs_pattern = re.compile(etfs_string)
filter_funds_pattern = re.compile(funds_string)
filter_top_pattern = re.compile(top_string)
        
for screen in all_quotes_screens:
    if (re.search(filter_non_us_pattern, screen)):    
        non_us_qscreen_list.append(screen)
    elif (re.search(filter_etfs_pattern, screen)):      
        etf_qscreen_list.append(screen)
    elif (re.search(filter_funds_pattern, screen)):     
        funds_qscreen_list.append(screen)    
    elif (re.search(filter_top_pattern, screen)):       
        top_qscreen_list.append(screen)
    else: 
        remaining_qscreen_list.append(screen)

filter_quotes_data_screens = non_us_qscreen_list + etf_qscreen_list + funds_qscreen_list + top_qscreen_list
```

```python
## creating and writing all qiotes non-us , non etf , non top data
##temp_file = r"C:\Users\pythonProject\data\data_produced\yquerry_screens\quotes_industry_data.csv"
quotes_columns = ['symbol','averageAnalystRating','regularMarketPrice','customPriceAlertConfidence',
                  'fiftyTwoWeekRange','sharesOutstanding','fiftyTwoWeekLow','fiftyTwoWeekHigh',
                  'regularMarketVolume',  'averageDailyVolume10Day', 'averageDailyVolume3Month',
                  'marketCap', 'bookValue', 'longName']

full_file_name = f"{formatted_date}_quotes_industry_data.csv"
output_file = os.path.join(base_folder_path, full_file_name)

# industry_data_screens quotes_data_screens
quotes_screen_values = []
quotes_industry_df = pd.DataFrame()
#filter_out_list = ['insurance_brokers', 'insurance_diversified', 'analyst_strong_buy_stocks',
#                   'infrastructure_operations','latest_analyst_upgraded_stocks']
filter_out_list = ['infrastructure_operations']
def build_quotes_industry_screen_data(name):
    global quotes_industry_df
    quotes_industry_data_dict = s.get_screeners([name])
    quotes_industry_data_list = quotes_industry_data_dict[name]['quotes']
    quotes_screen_values.extend([name] * len(quotes_industry_data_list))
    temp_df = pd.DataFrame(quotes_industry_data_list, columns=quotes_columns)
    quotes_industry_df = pd.concat([quotes_industry_df, temp_df])
    
    

# Display the DataFrame
for scr in all_quotes_screens:
    if scr not in filter_out_list:
        if (scr not in filter_quotes_data_screens):
            print(scr)
            build_quotes_industry_screen_data(scr)

# Assign the list to the "screen" column
quotes_industry_df["screen"] = quotes_screen_values
quotes_industry_df.set_index(keys=["screen"], inplace = True)
quotes_industry_df.to_csv(output_file)
```
