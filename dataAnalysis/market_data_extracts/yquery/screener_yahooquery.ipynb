```python
##'top_options_open_interest'
from datetime import datetime

# Function to convert Unix timestamp to formatted date
def convert_unix_to_date(timestamp):
    date_time = datetime.utcfromtimestamp(timestamp)
    formatted_date = date_time.strftime('%Y-%m-%d')
    return formatted_date

industry = 'top_options_open_interest'
columns = ['symbol', 'optionsType', 'strike', 'expireDate', 'openInterest', 'regularMarketPrice', 'fiftyTwoWeekRange', 'regularMarketVolume',
           'priceHint']
my_dict = s.get_screeners([industry])
my_data_list = my_dict[industry]['quotes']

top_options_open_interest_df = pd.DataFrame(my_data_list, columns=columns)
top_options_open_interest_df.set_index(keys=['symbol'], inplace = True)

top_options_open_interest_df['expireDate'] = top_options_open_interest_df['expireDate'].apply(convert_unix_to_date)
top_options_open_interest_df
```
